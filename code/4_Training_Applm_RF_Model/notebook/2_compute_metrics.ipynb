# Compute AUROC and AUPRC metrics using PRROC in R

```r
library(PRROC)
library(dplyr)
```

```r
# Ubuntu                    22.04

# miniconda3                24.1.2

# R version 4.3.3 (2024-02-29)
# dplyr_1.1.4
# PRROC_1.3.1
```

```r
df = read.csv('../2_results/ex1/test_labeled.csv')
roc <- roc.curve(scores.class0 = df$y_hat, weights.class0 = df$y, curve=TRUE)
pr <- pr.curve(scores.class0 = df$y_hat, weights.class0 = df$y, curve=TRUE, rand.compute=TRUE)
```

```r
plot(roc)
```

```r
plot(pr)
```
